                                                    # Expense Tracker Project

```python

import datetime
from itertools import count
import matplotlib.pyplot as plt
```

```python
#Dictionary of all the added expenses.
Tracker ={
    'expenses': []
}
```

```python

# Generates new id for added expense, by using count function from itertools module.
# count() will return consecutive integers starting from 0, the next() returns the next value from counter.
gen_id = lambda counter=count(): next(counter) 

#First feature Adding Expense
def addExpense(lst): # receives a list from the input of the user contains the info about the added expense.
    
    new_expns = {
        'id': gen_id(),
        'category': lst[0],
        'amount': float(lst[1]),
        'date': datetime.datetime.strptime(lst[2], "%Y-%m-%d").date(), 
        'description': lst[3]
    }

    Tracker['expenses'].append(new_expns)
    return new_expns

#Second feature Calculate Total Expenses, by calculating the total expenes of every month.   
def calTotal(): 
    months_tot={}
    for exp in Tracker['expenses']:
        m = exp['date'].strftime("%Y-%m")
        if m in months_tot:
            months_tot[m] += exp['amount']
        else:
            months_tot[m] = exp['amount']
    return months_tot

#Third feature Display All of the expenses details.
def viewDetails(): 

    #Check if there is no expenses to be viewed
    if Tracker['expenses'] == []:
        return print("|---------------------| \n No Expenses exists.\n|---------------------|")
    
    for exp in Tracker['expenses']:
        print("|-----------------------|")
        print(f" ID: {exp['id']}")
        print(f" Category: {exp['category']}")
        print(f" Amount: {exp['amount']}")
        print(f" Date: {exp['date']}")
        print(f" Descreption: {exp['description']}")
        
#Fourth feature Visualized report of each month expenses      
def report(): 
    
    months_tot = calTotal()

    #Check if there is no expenses to be reported
    if months_tot == {}:
        return print("|---------------------| \n No Expenses exists.\n|---------------------|")
    
    months = list(months_tot.keys())
    expenses = list(months_tot.values())

    # using matplotlib to show graph.
    plt.figure(figsize=(8, 8))
    
    plt.bar(months, expenses)
    plt.xlabel('Year-Month')
    plt.ylabel('Total Expenses')
    plt.title('Monthly Expense Report')
    plt.yticks(expenses) 
    plt.show()


```

```python
# Main function to run the Expense Tracker program.
def main():
    while True:
        print("|----Expense Tracker----|")
        print(" 1.Add Expense")
        print(" 2.Calculate Total Expenses")
        print(" 3.View Expense Details")
        print(" 4.Generate Expense Report")
        print(" 5.Exit")

        user_choice= input(" Select the number of the feature:")

        # First feature----------------------------------------------
        if user_choice =="1":
        
            print("|----Fill The details of the Expesnse----|")
            category = input(f" Enter the Category: ")

            while True: #To keep checking if the amount is valid
                Entered_amount = input(f" Enter the Amount: ")
                try:
                    amount = float(Entered_amount)
                    break
                except:
                    print(" Invalid amount, you must Enter numbers only.")

            date = input(f" Enter the date (YYYY-MM-DD): ")
            description = input(f"  Enter the Descreption: ")

            lst = [category , amount, date ,description]
            new_expns = addExpense(lst)

            print("|__________________________________________|")
            print(f" Category: {new_expns['category']}")
            print(f" Amount: {new_expns['amount']}")
            print(f" Date: {new_expns['date']}")
            print(f" Descreption: {new_expns['description']}")
            print("|__________________________________________|")
            print(" Expense have been added successfully!")
            continue

            

        # Second feature---------------------------------------------    
        elif user_choice =="2":
            print('|----Total Month Expenses----|')
            tot_exp = calTotal()
            for month , total in tot_exp.items():
                print("|---------------------------|")
                print(f" Month: {month}")
                print(f" Total Expenses: {total}")

        # Third feature------------------------------------------------
        elif user_choice =="3":  
            print("|----All Expenses---|")
            viewDetails()
            

        # Fourth feature------------------------------------------------
        elif user_choice =="4":  
            report()

        elif user_choice =="5":
            print(" Existing...")
            break
        else:
            print(" Invalid choice, please select from 1-5.")
                


if __name__ == "__main__":
    main()

            
    


```
